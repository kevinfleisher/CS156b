```python
import numpy as np
import pandas as pd
```

```python
file_path = '/Users/omidgolami/CS156b/Data/train.csv'
train_df = pd.read_csv(file_path)
train_df
```

```python
ratios = []
for col in train_df.columns[6:20]:
    num_nan = train_df[col].isna().sum()
    num_zeros = (train_df[col] == 0).sum()
    num_ones = (train_df[col] == 1).sum()
    num_minus_ones = (train_df[col] == -1).sum()
    total_rows = len(train_df)
    ratio_nan = num_nan / total_rows
    ratio_zero = num_zeros / total_rows
    ratio_ones = num_ones / total_rows
    ratio_minus_ones = num_minus_ones / total_rows
    ratios.append({
        'Column': col,
        'NaN Ratio': ratio_nan,
        '-1 Ratio': ratio_minus_ones,
        '0 Ratio': ratio_zero,
        '1 Ratio': ratio_ones
    })

ratios_df = pd.DataFrame(ratios)

ratios_df
```
